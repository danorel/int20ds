### Preprocessing

1. Finding average speed during the running time

```python
speed_mean_train_df = nodes_train_df.groupby(["Id"], as_index=False).agg(**{"average_speed": ("speed", "mean")})
```

### Big table for train

```python
big_table_train_df = orders_train_df \
    .merge(nodes_train_df, on='Id', how='inner') \
    .merge(speed_mean_train_df, on='Id', how='inner') \
    .merge(geolocation_train_df, left_on=['Id', 'node_start', 'node_finish'], right_on=["Id", 'node_start', 'node_finish'], how='left')
```

### Big table for final test

```python
orders_test_df = pd.read_csv("datasets/csv/orders_test.csv")
nodes_test_df = pd.read_csv("datasets/csv/nodes_test.csv")
geolocation_test_df = pd.read_csv("datasets/csv/geolocation_test.csv")
```

```python
geolocation_test_df = geolocation_test_df[["Id", "node_start", "node_finish", "lon_start", "lat_start", "lon_end", "lat_end"]]
```

```python
big_table_test_df = orders_test_df.merge(nodes_test_df, on='Id', how='inner')
```

```python
speed_mean_test_df = nodes_test_df.groupby(["Id"], as_index=False).agg(**{"average_speed": ("speed", "mean")})
```

```python
big_table_test_df = big_table_test_df \
    .merge(speed_mean_test_df, on='Id', how='inner') \
    .merge(geolocation_test_df, left_on=['Id', 'node_start', 'node_finish'], right_on=["Id", 'node_start', 'node_finish'], how='left')
```

```python
big_table_test_df
```

```python
big_table_test_df.to_csv("datasets/big-table-test.csv", index=False)
```
